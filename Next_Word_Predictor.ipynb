```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
filepath = "/content/drive/MyDrive/DeepLearning/data.txt"
```

Read data from file , Split into sentences and assign numbers to each unique word using tokenizer

```python
data = ""
with open(filepath , 'r' , encoding = "utf-8") as f:
    data = f.read()

my_list = data.split('.')
my_list = [sen for sen in my_list if len(sen) > 2]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(my_list)
tokenizer.word_index
```

convert english sentences into numbers and save in a list and make features by adding one more word

```python
input_seq=[]
print(len(my_list))
for i in my_list:
    tokenized_sentence=tokenizer.texts_to_sequences([i])[0]
    # print(tokenized_sentence[0])

    for i in range(1,len(tokenized_sentence)):
        input_seq.append(tokenized_sentence[:i+1])
```

```python
# print(len(input_seq))
# print(input_seq)
input_seq
```

```python
max_len = max([len(x) for x in input_seq])
max_len
```

Add padded zeros to make size of features same for seupervised learning

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input = pad_sequences(input_seq , maxlen = max_len , padding = 'pre')
padded_input
```

  Separate input and output from padded vector

```python
X = padded_input[:,:-1]
```

```python
Y = padded_input[:,-1]
```

```python
Y
```

```python
print(X.shape)
print(Y.shape)
```

One hot encoding of Y accoroding to no of words unique in my data (tokenizer)

one hot encoding starts from 0 thats why 1 added to length

```python
from tensorflow.keras.utils import to_categorical
Y = to_categorical(Y , num_classes = len(tokenizer.word_index)+1)
```

```python
Y.shape
```

```python
import numpy as np
Y = np.argmax(Y, axis=1)
```

```python
Y.shape
```

```python
Y[:4]
```

Build architecture of model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding , LSTM , Dense
```

Embedding layer ka main faida ye hai ke wo words ko numbers se vectors me convert karta hai — aur wo vectors meaningful relationships sikhtay hain.

Dense layer os fro output calulation based on activation function as it is multiclass classification

```python
model = Sequential()
model.add(Embedding(8929 , 100 ))
model.add(LSTM(150)) #no of nodes in every gate
model.add(Dense(8929 , activation = 'softmax')) #no of words in vocab as need prob of each word in prob to give answer
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.build(input_shape=(None, 104))
```

```python
model.summary()
```

```python
X
```

```python
model.fit(X , Y , epochs = 100)
```
